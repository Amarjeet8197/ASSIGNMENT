```python
# Question-1
#You are given a binary tree. The binary tree is represented using the TreeNode class. Each TreeNode has an integer value and left and right children, represented using the TreeNode class itself. Convert this binary tree into a binary search tree.
#Input:

#        10

#    /   \

 #    2      7

#   /   \

# 8      4

#Output:

#        8

#      /   \

#    4     10

#  /   \

#2      7
```

```python
class TreeNode:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None

def inorderTraversal(root):
    if root is None:
        return []

    result = []
    stack = []
    current = root

    while current is not None or stack:
        while current is not None:
            stack.append(current)
            current = current.left

        current = stack.pop()
        result.append(current.val)
        current = current.right

    return result

def convertToBST(root):
    if root is None:
        return

    values = inorderTraversal(root)
    values.sort()

    index = 0
    stack = []
    current = root

    while current is not None or stack:
        while current is not None:
            stack.append(current)
            current = current.left

        current = stack.pop()
        current.val = values[index]
        index += 1
        current = current.right

    return root

# Construct the binary tree
root = TreeNode(10)
root.left = TreeNode(2)
root.right = TreeNode(7)
root.left.left = TreeNode(8)
root.left.right = TreeNode(4)

# Convert the binary tree to a binary search tree
convertToBST(root)

# Perform inorder traversal to check the result
def inorderTraversalBST(root):
    if root is None:
        return []

    result = []
    stack = []
    current = root

    while current is not None or stack:
        while current is not None:
            stack.append(current)
            current = current.left

        current = stack.pop()
        result.append(current.val)
        current = current.right

    return result

# Print the converted binary search tree
print(inorderTraversalBST(root))
```

```python
# Question-3:
#Write a program to convert a binary tree to a doubly linked list.
#Input:

#        10

#       /   \

#     5     20

#         /   \

 #       30     35

#Output: 5 10 30 20 35
```

```python
class TreeNode:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None

def binaryTreeToDLL(root):
    if root is None:
        return None

    # Convert the left subtree to a DLL
    left_head = binaryTreeToDLL(root.left)

    # Convert the right subtree to a DLL
    right_head = binaryTreeToDLL(root.right)

    # Make the root a single-node DLL
    root.left = root
    root.right = root

    # Merge the left DLL, root DLL, and right DLL
    left_tail = left_head.left if left_head else None
    right_tail = right_head.left if right_head else None

    if left_tail:
        left_tail.right = root
        root.left = left_tail
        left_head.left = right_tail if right_tail else root

    if right_head:
        right_head.left = root
        root.right = right_head
        right_tail.right = left_head if left_head else root

    return left_head if left_head else root

def printDLL(head):
    current = head
    while current:
        print(current.val, end=" ")
        current = current.right
        if current == head:
            break
    print()

# Construct the binary tree
root = TreeNode(10)
root.left = TreeNode(5)
root.right = TreeNode(20)
root.right.left = TreeNode(30)
root.right.right = TreeNode(35)

# Convert the binary tree to a doubly linked list
head = binaryTreeToDLL(root)

# Print the doubly linked list
printDLL(head)
```

```python
# Question-4:

#Write a program to connect nodes at the same level.

#Input:         1

 #     /   \

 #   2      3

#  /   \   /   \

#4     5 6    7

#Output:
#1 → -1 2 → 3 3 → -1 4 → 5 5 → 6 6 → 7 7 → -1
```

```python
from collections import deque

class Node:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None
        self.nextRight = None

def connectNodes(root):
    if root is None:
        return None

    queue = deque()
    queue.append(root)

    while queue:
        levelSize = len(queue)

        for i in range(levelSize):
            node = queue.popleft()

            if i < levelSize - 1:
                node.nextRight = queue[0]

            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)

        node.nextRight = None

    return root

def printConnectedNodes(root):
    current = root
    while current:
        print(current.val, end=" → ")
        current = current.nextRight
    print("-1")
```

```python
# Construct the binary tree
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
root.right.left = Node(6)
root.right.right = Node(7)

# Connect nodes at the same level
connectNodes(root)

# Print the connected nodes at each level
printConnectedNodes(root.left.left)
printConnectedNodes(root.left)
printConnectedNodes(root.right.left)
printConnectedNodes(root.right)
```
