```python
#Question-2
#Construct the BST (Binary Search Tree) from its given level order traversal.
#Example:
#Input: arr[] = {7, 4, 12, 3, 6, 8, 1, 5, 10}
#Output: BST:

#            7

#         /    \

#       4     12

#     /  \     /

#    3   6  8

#   /    /     \

# 1    5      10
```

```python
# Definition of a binary tree node
class TreeNode:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None

def constructBST(levelOrder):
    if not levelOrder:
        return None

    # Create the root node
    root = TreeNode(levelOrder[0])
    levelOrder.pop(0)

    queue = [root]

    while levelOrder:
        currentNode = queue.pop(0)

        # Find the index of the first element greater than currentNode
        index = 0
        while index < len(levelOrder) and levelOrder[index] <= currentNode.val:
            index += 1

        # Split the level order array into left and right subtrees
        leftSubtree = levelOrder[:index]
        rightSubtree = levelOrder[index:]

        if leftSubtree:
            currentNode.left = TreeNode(leftSubtree[0])
            queue.append(currentNode.left)
            leftSubtree.pop(0)

        if rightSubtree:
            currentNode.right = TreeNode(rightSubtree[0])
            queue.append(currentNode.right)
            rightSubtree.pop(0)

        # Update the level order array
        levelOrder = leftSubtree + rightSubtree

    return root

# Test case
levelOrder = [7, 4, 12, 3, 6, 8, 1, 5, 10]
root = constructBST(levelOrder)

# Function to print the BST in inorder traversal
def inorderTraversal(node):
    if node:
        inorderTraversal(node.left)
        print(node.val, end=" ")
        inorderTraversal(node.right)

# Print the constructed BST
inorderTraversal(root)
```

```python
#Question-3

#Given an array of size n. The problem is to check whether the given array can represent the level order traversal of a Binary Search Tree or not.

#Examples: Input1 : arr[] = {7, 4, 12, 3, 6, 8, 1, 5, 10}

#Output1 : Yes For the given arr[], the Binary Search Tree is:

#            7

#         /    \

 #      4     12

 #    /  \     /

  #  3   6  8

#   /    /     \

# 1    5      10
#Input2 : arr[] = {11, 6, 13, 5, 12, 10}
#Output2 : No
#The given arr[] does not represent the level order traversal of a BST.
```

```python
def isValidLevelOrder(arr):
    if not arr:
        return True

    n = len(arr)
    i = 0

    while i < n:
        rightIndex = -1

        # Find the first element greater than arr[i]
        for j in range(i+1, n):
            if arr[j] > arr[i]:
                rightIndex = j
                break

        # Check if there is any smaller element after rightIndex
        for k in range(rightIndex+1, n):
            if arr[k] < arr[i]:
                return False

        # Update i value to rightIndex
        i = rightIndex

    return True

# Test cases
arr1 = [7, 4, 12, 3, 6, 8, 1, 5, 10]
print("Output1:", isValidLevelOrder(arr1))  # True

arr2 = [11, 6, 13, 5, 12, 10]
print("Output2:", isValidLevelOrder(arr2))  # False
```
